```python
# selected features list
features = ['index' , 'title' , 'genres' , 'keywords' , 'tagline' , 'cast' , 'director']
```

```python
sns.heatmap(df.isnull())
```

```python
df = df[features]
df
```

```python
sns.heatmap(df.isnull())
```

```python
for f in features:
    df[f] = df[f].fillna('')
```

```python
df.dtypes
```

```python
# combined feature
def combine_features(record):
    return record['title'] + " " + record['keywords']+" "+record['cast']+" "+record['genres'] + " "+record['tagline']+" "+record['director']
```

```python
df["combined_features"] = df.apply(combine_features , axis = 1)
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(df['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(count_matrix)
```

```python
movie = 'Iron Man'
Movie = movie
```

```python
# now recommending movies
```

```python
def movie_index(movie_name):
    return df[df.title == movie_name]["index"].values[0]
```

```python
movieIndex = movie_index(movie)
print(movieIndex)
```

```python
similar_movies = list(enumerate(similarity[movieIndex]))
```

```python
similar_movies = sorted(similar_movies , key = lambda x : x[1] , reverse = True )
similar_movies
```

```python
def get_title(index):
    return df[df.index == index]["title"].values[0]
```

```python
 lim = 0 
result = list()
    
for movie in similar_movies:
   if(lim > 0):
    result.append(get_title(movie[0]))
   if(lim == 10 ) :
        break
   lim+=1
```

```python
print("Suggested movies to your choice \"{}\" are :\n".format(Movie))

for movie in result:
   print(movie)
   print("\n")
```
